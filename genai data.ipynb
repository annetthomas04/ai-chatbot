```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/Annet Thomas/Downloads/genai.csv")
```

```python
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
```

```python
for company in df['Company'].unique():
    company_data = df[df['Company'] == company]
    plt.plot(company_data['Date'], company_data['Total Revenue'], label=company)

plt.title('Total Revenue by Company Over Years')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.legend()
plt.show()
```

```python
df['ROA (%)'] = (df['Net Income'] / df['Total Assets']) * 100
```

```python
df['Liquidity Ratio'] = df['Total Assets'] / df['Total Liabilities']
```
